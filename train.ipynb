```python
import os
import subprocess
import sys

def run(cmd, check=True):
    """Run shell command with live output"""
    result = subprocess.run(cmd, shell=True, text=True)
    if check and result.returncode != 0:
        raise RuntimeError(f'Command failed: {cmd}')
    return result

# ── System packages ───────────────────────────────────────────
print('Installing system dependencies...')
run('apt-get update -qq')
run('apt-get install -y -qq autoconf automake libtool gcc make git sox ffmpeg wget unzip')

# ── Python packages ───────────────────────────────────────────
# Install one by one so a single failure doesn't block everything
print('Installing Python packages...')

packages = [
    'pesq',
    'pystoi',
    'soundfile',
    'librosa',
    'tqdm',
]

for pkg in packages:
    print(f'  Installing {pkg}...')
    result = subprocess.run(
        [sys.executable, '-m', 'pip', 'install', '-q', pkg],
        capture_output=True, text=True
    )
    if result.returncode != 0:
        print(f'  WARNING: {pkg} failed — {result.stderr.strip()}')
    else:
        print(f'  {pkg} OK')

# ── Verify GPU ────────────────────────────────────────────────
print('\nVerifying GPU...')
import tensorflow as tf

gpus = tf.config.list_physical_devices('GPU')
print(f'TensorFlow version : {tf.__version__}')
print(f'GPUs detected      : {len(gpus)}')
for g in gpus:
    print(f'  {g}')

if len(gpus) == 0:
    print('WARNING: No GPU detected!')
    print('Go to Settings -> Accelerator -> GPU T4 x2 and restart session.')
else:
    print('GPU setup confirmed!')
```

```python
import os
import struct
import numpy as np
import librosa
from datasets import load_dataset

SPEECH_RAW = "/kaggle/working/speech.raw"

LANGUAGES = [
    "SPRINGLab/IndicTTS-Hindi",
    "SPRINGLab/IndicTTS-English",
]

MAX_PER_LANG = 1500

# Remove old raw file
if os.path.exists(SPEECH_RAW):
    os.remove(SPEECH_RAW)

total_saved = 0

with open(SPEECH_RAW, "ab") as raw_out:

    for lang_repo in LANGUAGES:

        lang_name = lang_repo.split("/")[-1]
        print(f"\nDownloading {lang_name}...")

        try:
            ds = load_dataset(
                lang_repo,
                split="train",
                streaming=True,
                trust_remote_code=True
            )

            saved = 0

            for sample in ds:

                if saved >= MAX_PER_LANG:
                    break

                audio = sample["audio"]["array"]
                sr    = sample["audio"]["sampling_rate"]

                # mono
                if audio.ndim > 1:
                    audio = audio.mean(axis=1)

                # resample
                if sr != 48000:
                    audio = librosa.resample(
                        audio.astype(np.float32),
                        orig_sr=sr,
                        target_sr=48000
                    )

                # normalize safety
                audio = np.clip(audio, -1.0, 1.0)

                # float -> int16 PCM
                pcm16 = (audio * 32767).astype(np.int16)

                # append directly
                raw_out.write(pcm16.tobytes())

                saved += 1
                total_saved += 1

                if saved % 200 == 0:
                    print(f"    {lang_name}: {saved}/{MAX_PER_LANG}")

            print(f"Done: {saved}")

        except Exception as e:
            print(f"ERROR: {e}")

print(f"\nTOTAL FILES: {total_saved}")
print(f"RAW FILE SAVED TO: {SPEECH_RAW}")
```

```python
import os
import glob
import subprocess

NOISE_RAW = "/kaggle/working/noise.raw"
MY_NOISE_INPUT = "/kaggle/input/datasets/vishalhariharan97513/noisy-data/noise"

# Delete old raw
if os.path.exists(NOISE_RAW):
    os.remove(NOISE_RAW)

noise_files = glob.glob(os.path.join(MY_NOISE_INPUT, "*.mp3"))

print(f"Found {len(noise_files)} noise files")

with open(NOISE_RAW, "ab") as outfile:

    for src in sorted(noise_files):

        name = os.path.basename(src)
        print(f"Converting: {name}")

        # ffmpeg outputs raw PCM directly to stdout
        cmd = [
            "ffmpeg",
            "-i", src,
            "-ar", "48000",
            "-ac", "1",
            "-f", "s16le",
            "-acodec", "pcm_s16le",
            "-"
        ]

        proc = subprocess.run(
            cmd,
            stdout=subprocess.PIPE,
            stderr=subprocess.DEVNULL
        )

        outfile.write(proc.stdout)

print("\nnoise.raw created successfully")

size_gb = os.path.getsize(NOISE_RAW) / 1e9
print(f"RAW size: {size_gb:.2f} GB")
```

```python
import numpy as np
import h5py

INPUT  = "/kaggle/working/features.f32"
OUTPUT = "/kaggle/working/training.h5"

# Load raw floats
data = np.fromfile(INPUT, dtype=np.float32)

print("Original floats:", len(data))

# RNNoise frame size
COLS = 87

# Trim excess
usable = (len(data) // COLS) * COLS
data = data[:usable]

rows = usable // COLS

print("Usable floats:", usable)
print("Rows:", rows)
print("Cols:", COLS)

# Reshape
data = data.reshape(rows, COLS)

# Save HDF5
with h5py.File(OUTPUT, "w") as hf:
    hf.create_dataset("data", data=data)

print("\ntraining.h5 created successfully")
```

```python
import torch, os
import matplotlib.pyplot as plt

ckpt_dir = "/kaggle/working/output/checkpoints"
files = sorted(os.listdir(ckpt_dir))

epochs = []
losses = []

for f in files:
    path = os.path.join(ckpt_dir, f)
    ckpt = torch.load(path, map_location='cpu')
    epochs.append(ckpt['epoch'])
    losses.append(ckpt['loss'])

plt.figure(figsize=(12, 5))
plt.plot(epochs, losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("RNNoise Training Loss")
plt.grid(True)
plt.show()

# Print best epoch
best_epoch = epochs[losses.index(min(losses))]
print(f"Best epoch : {best_epoch}")
print(f"Best loss  : {min(losses):.5f}")
print(f"Last loss  : {losses[-1]:.5f}")
```
